```python
country = pd.read_csv('./countries.csv')
state = pd.read_csv('./states.csv')
cities = pd.read_csv('./cities.csv')
```

```python
countryDF = pd.DataFrame(country)
stateDF = pd.DataFrame(state)
citiesDF = pd.DataFrame(cities)
```

```python
indianStates = stateDF.loc[stateDF['country_id'] == 101]
```

```python
indianCities = citiesDF.loc[citiesDF['country_id'] == 101]
```

```python
indianCities.head()
```

```python
indianCities.describe()
```

```python
indianData = pd.merge(indianCities, indianStates,
                      left_on="state_id", right_on="id")
```

```python
indianData.head()
```

```python
statesObj = {}
for name, group in indianData.groupby('state_name'):
    citiesObj = group["name_x"].tolist()
    statesObj[name] = {'cities': citiesObj}


```

```python
with open("states.json", "w") as f:
    json.dump(statesObj, f, indent=4)
```
